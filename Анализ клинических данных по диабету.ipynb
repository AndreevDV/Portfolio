### Справочные данные

__Гипертония__ - заболевание сердечно-сосудистой системы, проявляющееся в повышенном артериальном давлении (АД ≥ 140/90 мм.рт.ст.). Является одним из факторов риска развития сахарного диабета

__Болезни сердца__ (ИБС в частности) - еще один фактор влияющий на развитие диабета

__Курение__ - официального доказательства влияния курения на развитие сахарного диабета нет, однако курение в случае уже поставленного диагноза может сильно понижать эффективность лечения и вызвать осложнения, способствующие летальному исходу.

__Индекс массы тела (ИМТ)__ - величина, позволяющая оценить степень соответствия массы человека и его роста и тем самым косвенно судить о том, является ли масса недостаточной, нормальной или избыточной.

Формула для расчёта ИМТ:

    ИМТ = масса тела(кг) / (рост(м)^2)

Для оценки ИМТ у взрослых ВОЗ использует следующие критерии:

- менее 18,5 кг/м2 — дефицит массы тела
- 18,5–24,9 кг/м2 — здоровый вес
- 25–29,9 кг/м2 — избыточная масса тела
- 30–34,9 кг/м2 — ожирение 1 степени
- 35–39,9 кг/м2 — ожирение 2 степени
- более 40 кг/м2 — ожирение 3 степени

__Уровень гликолизированного гемоглобина(hbA1c)__ - биохимический показатель, отражающий усредненное содержание глюкозы в крови за предшествующие 3 месяца.

При различных заболеваниях происходит химическое изменение структуры гемоглобина. Так, при сахарном диабете к гемоглобину присоединяется глюкоза с образованием гликированного гемоглобина. И чем выше уровень глюкозы в крови, тем больше гемоглобина модифицируется в гликированный гемоглобин.

Образование гликированного гемоглобина необратимо (то есть гликированный гемоглобин не может обратно стать нормальным гемоглобином), и его концентрация в крови зависит как от продолжительности жизни эритроцитов (до 120 дней), так и от концентрации глюкозы в крови. Поскольку скорость образования гликированного гемоглобина тем выше, чем выше уровень глюкозы в крови, то по его концентрации можно судить о среднем содержании глюкозы в крови за последние 8–12 недель перед исследованием.

Одной из главной причин повышения уровня hbA1c является сахарный диабет, при этом нормализация уровня глюкозы у пациентов с диабетом свидетельствует об эффективном лечении.

Итерпретация уровня hbA1c:

- менее 6.0% - нормальный уровень
- 6,0% – 6,5% (5,7% – 6,4%) - повышенный риск развития диабета и его осложнений по данным ВОЗ (по данным ADA, American Diabetes Association)
- более 6,5% - диагностический критерий сахарного диабета

__Уровень глюкозы в крови__ - биохимическое исследование, которое используют для диагностики и контроля сахарного диабета и других заболеваний, связанных с нарушением обмена углеводов.

Интерпретация уровня глюкозы в крови:

- менее 3.3 ммоль/л - гипогликемия
- 3.3 - 5.5 ммоль/л - нормальный уровень
- 5.5 - 8.3 ммоль/л - гипергликемия
- более 8.3 ммоль/л - диагностический критерий диабета

У пациентов с сахарным диабетом может отмечаться низкий уровень глюкозы в следующих ситуациях:

- передозировка инсулина
- передозировка препаратов, снижающих уровень глюкозы в крови
- повышенная физическая нагрузка
- пропуск приема пищи
- прием алкоголя

Повышенный уровень глюкозы в крови сигнализирует о предиабетическом состоянии

```python
# функция для категоризации уровня гликолизированного гемоглобина у пациентов
def categorize_age(age):
    if age < 18:
        return 'ребенок'
    elif 18 <= age < 60:
        return 'взрослый'
    elif age >= 60:
        return 'пожилой'
```

```python
# функция для категоризации пациентов по ИМТ
def categorize_bmi(bmi):
    if bmi < 18.5:
        return 'дефицит массы тела'
    elif 18.5 <= bmi < 25:
        return 'норма'
    elif 25 <= bmi < 30:
        return 'избыточный вес'
    elif 30 <= bmi < 40:
        return 'ожирение'
    elif bmi >= 40:
        return 'крайнее ожирение'
```

```python
# функция для категоризации уровня гликолизированного гемоглобина у пациентов
def categorize_hbA1c(hbA1c):
    if hbA1c < 6:
        return 'нормальный'
    elif 6 <= hbA1c < 6.5:
        return 'повышенный'
    elif hbA1c >= 6.5:
        return 'диагностический критерий диабета'
```

```python
# функция для категоризации уровня глюкозы в крови у пациентов
def categorize_glucose(glucose):
    if glucose < 61:
        return 'гипогликемия'
    elif 61 <= glucose < 100:
        return 'нормальный уровень'
    elif 100 <= glucose < 125:
        return 'повышенный'
    elif glucose >= 125:
        return 'диагностический критерий диабета'
```

```python
# применение ранее созданных функций
data['age_category'] = data['age'].apply(categorize_age)
data['bmi_category'] = data['bmi'].apply(categorize_bmi)
data['hbA1c_category'] = data['hbA1c_level'].apply(categorize_hbA1c)
data['glucose_category'] = data['blood_glucose_level'].apply(categorize_glucose)
```

### Построение моделей машинного обучения для предсказывания диабета

#### Стандартизация данных

```python
# разделим наши данные на признаки (матрица X) и целевую переменную (y)
X = data.drop(['year',
                  'diabetes',
                  'location', 
                  'race', 
                  'smoking_history', 
                  'bmi_category', 
                  'hbA1c_category', 
                  'glucose_category',
                  'age_category'], axis=1)
y = data['diabetes']

X = pd.get_dummies(X)

# разделяем модель на обучающую и валидационную выборку
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)

# обучение StandartScaler на обучающей выборке
scaler = StandardScaler()
scaler.fit(X_train)

# преобразование обучающий и валидационный наборы данных
X_train_st = scaler.transform(X_train)
X_test_st = scaler.transform(X_test)
```

#### Обучение моделей на train-выборке

```python
# задание алгоритма для модели логистической регрессии
lr_model = LogisticRegression(random_state=0)
# обучение модели
lr_model.fit(X_train_st, y_train)
```

```python
# задание алгоритма для новой модели на основе алгоритма решающего дерева
tree_model = DecisionTreeClassifier(random_state=0)
# обучение модели решающего дерева
tree_model.fit(X_train_st, y_train)
```

```python
# задание алгоритма для новой модели на основе алгоритма случайного леса
rf_model = RandomForestClassifier(n_estimators = 100, random_state = 0) 
# обучение модели случайного леса
rf_model.fit(X_train_st, y_train)
```

```python
# задание алгоритма для новой модели на основе алгоритма градиентного бустинга
gb_model = GradientBoostingClassifier(n_estimators = 100, random_state = 0)
# обучение модели градиентного бустинга
gb_model.fit(X_train_st, y_train)
```

#### Прогнозирование заболевания

```python
# модель логистической регрессии
lr_predictions = lr_model.predict(X_test_st)
lr_probabilities = lr_model.predict_proba(X_test_st)[:, 1]
# модель на основе алгоритма решающего дерева
tree_predictions = tree_model.predict(X_test_st)
tree_probabilities = tree_model.predict_proba(X_test_st)[:, 1]
# модель на основе алгоритма случайного леса
rf_predictions = rf_model.predict(X_test_st)
rf_probabilities = rf_model.predict_proba(X_test_st)[:,1]
# модель на основе алгоритма градиентного бустинга
gb_predictions = gb_model.predict(X_test_st)
gb_probabilities = gb_model.predict_proba(X_test_st)[:,1]
```

#### Анализ результатов

```python
# определение функции, которая будет выводить метрики
def print_all_metrics(y_true, y_pred, y_proba, title='Метрики классификации'):
    print(title)
    print('\tAccuracy: {:.2f}'.format(accuracy_score(y_true, y_pred)))
    print('\tPrecision: {:.2f}'.format(precision_score(y_true, y_pred)))
    print('\tRecall: {:.2f}'.format(recall_score(y_true, y_pred)))
    print('\tF1: {:.2f}'.format(f1_score(y_true, y_pred)))
    print('\tROC_AUC: {:.2f}'.format(roc_auc_score(y_true, y_proba)))
```
